# Notebook 1 : Data Cleaning & Preparation  
## Customer Churn Capstone Project

This notebook focuses on preparing the **raw customer churn dataset**  
for accurate analysis and modeling in later stages.

### Objectives
- Load raw dataset  
- Inspect structure and data quality  
- Handle missing values and duplicates  
- Ensure correct data types  
- Save a cleaned dataset for further analysis


```python
import pandas as pd
import numpy as np
```

## 1.Load Raw Dataset

We begin by loading the original **customer churn dataset**  
to inspect its structure and quality.


```python
df = pd.read_csv("raw_data.csv")
df.head()
```

```python
df.info()
df.shape
```

## 2.Missing Value Analysis

Checking for null or missing values ensures  
the dataset is reliable for statistical analysis.


```python
df.isnull().sum()
```

## 3.Handling Missing Values

If missing values exist:
- Numerical columns → filled with median  
- Categorical columns → filled with mode  

This prevents bias during modeling.

```python
for col in df.columns:
    if df[col].dtype in ["int64","float64"]:
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna(df[col].mode()[0])
```

## 4.Duplicate Record Removal

Duplicate rows can distort statistical results  
and must be removed before analysis.


```python
duplicates = df.duplicated().sum()
print("Duplicate rows:",duplicates)

df = df.drop_duplicates()
```

## 5.Data Type Verification

Ensuring numerical columns are stored correctly  
prevents calculation and modeling errors.


```python
numeric_cols = ["Tenure","MonthlyCharges","TotalCharges"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col],errors="coerce")

df[numeric_cols].info()
```

## 6.Final Validation

Confirming:
- No missing values remain  
- Dataset shape is correct  
- Data is ready for analysis


```python
print("Missing values after cleaning : \n",df.isnull().sum())
print("\nFinal dataset shape : ",df.shape)
```

## 7.Conclusion

This notebook successfully prepared the raw data by:
- Handling missing values  
- Removing duplicates  
- Validating data types  
- Saving a clean dataset  

The dataset is now **ready for EDA and statistical analysis**  
in the next notebook.
