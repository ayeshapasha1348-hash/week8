# Notebook 2 : Exploratory Data Analysis (EDA)  
## Customer Churn Capstone Project

This notebook explores the **cleaned customer churn dataset**  
to understand patterns,relationships,and risk factors related to churn.

### Objectives
- Analyze churn distribution  
- Study tenure and pricing behavior  
- Examine contract based retention  
- Explore correlations between variables  
- Generate insights for statistical modeling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## 1.Load Cleaned Dataset

We load the **cleaned dataset** prepared in Notebook 1  
to begin exploratory analysis.


```python
df = pd.read_csv("cleaned_data.csv")
df.head()
```

```python
df.shape
df.describe()
```

## 2.Churn Distribution

Understanding how many customers **stayed vs churned**  
provides the baseline business picture.

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Churn",data=df)
plt.title("Customer Churn Distribution")
plt.xlabel("Churn(0 = No,1 = Yes)")
plt.ylabel("Customer Count")
plt.show()
```

### Insight
This chart reveals the **overall churn proportion**,  
helping estimate business risk and retention performance.


## 3.Tenure vs Churn

Customer tenure indicates **how long a customer stayed**  
and is a strong indicator of loyalty.


```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Churn",y="Tenure",data=df)
plt.title("Tenure Distribution by Churn")
plt.xlabel("Churn")
plt.ylabel("Tenure (Months)")
plt.show()
```

### Insight
Customers with **shorter tenure** are more likely to churn,  
showing high risk during early lifecycle stages.


## 4.Monthly Charges vs Churn

Pricing pressure may influence customer retention.


```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Churn",y="MonthlyCharges",data=df)
plt.title("Monthly Charges by Churn")
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")
plt.show()
```

### Insight
Customers paying **higher monthly charges** show  
a **greater tendency to churn**,indicating pricing sensitivity.

## 5.Contract Type vs Churn

Contract duration strongly affects retention behavior.


```python
plt.figure(figsize=(7,5))
sns.countplot(x="Contract",hue="Churn",data=df)
plt.title("Contract Type vs Churn")
plt.xticks(rotation=20)
plt.show()
```

### Insight
**Month-to-month contracts** show the highest churn,  
while **long-term contracts** improve retention stability.


## 6.Correlation Between Numerical Variables

Correlation helps identify **relationships influencing churn**.


```python
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Insight
- **Tenure and Monthly Charges** influence churn behavior.  
- These variables will be important for **statistical testing and modeling**.


## 7.Distribution of Key Numerical Features


```python
df[["Tenure","MonthlyCharges","TotalCharges"]].hist(figsize=(10,5),bins=20)
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

## 8.Key Findings from EDA

- Significant proportion of customers churn.  
- **Low-tenure customers** have highest churn risk.  
- **Higher monthly charges** increase churn likelihood.  
- **Month-to-month contracts** show weakest retention.  
- Tenure and charges are strong predictors for modeling.
